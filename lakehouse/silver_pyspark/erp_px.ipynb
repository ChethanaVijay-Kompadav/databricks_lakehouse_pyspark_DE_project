# Read Source Table

```python
df = spark.table('workspace.bronze_pyspark.erp_px_cat')
df.display()
```

# Import methods & libraries

```python
from pyspark.sql.functions import *
from pyspark.sql.functions import col, when, concat, coalesce, trim, lit, substring, regexp_replace
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window
```

# Explore source table

```python
(df.select(col('CAT')).distinct()).show()
```

```python
(df.select(col('SUBCAT')).distinct()).show()
```

```python
(df.select(col('MAINTENANCE')).distinct()).show()
```

```python
px_cat= (df.select(col('ID').alias('category_id'),
           col('CAT').alias('category'),
           col('SUBCAT').alias('subcategory'),
           col('MAINTENANCE').alias('maintenance')
           
           ))
```

# Drop Target table if it already exists

```python
spark.sql("""DROP TABLE IF EXISTS workspace.silver_pyspark.erp_px_cat""")
```

# Create Target table & load the transformed data into it

```python
px_cat.write.format('delta').mode('overwrite').saveAsTable('workspace.silver_pyspark.erp_px_cat')
```
